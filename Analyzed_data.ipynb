```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('car sales.csv')
print(df.shape)
```

```python
df.drop_duplicates()
```

--> EXTRACT UNIQUE VALUE OF CAR AND YEAR

```python
unique_yrs = df['year'].unique()
unique_car = df['car'].unique()
print(unique_yrs)
print(unique_car)
```

--> CALCULATE MAX PRICE

```python
df = df.dropna(subset=['price','car'])
max_price = df[('p'
                'rice')].max()
print(max_price)
```

```python
highpricecar = df[df['price'] == max_price]
print(highpricecar)
```

```python
min_price = 5000
max_price = 2500000
car = df[(df['price'] >= min_price ) & (df['price'] <= max_price)]['car'].unique()
print(car)
```

```python
df['engV'] = df['engV'].replace(" ", "0")
```

```python
print(df['engV'])
```

```python
plt.figure(figsize=(6, 6))
sns.histplot(df['price'], kde=True, bins=50, color='skyblue')
plt.title('Distribution of Car Prices')
plt.xlabel('Frequency')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

Top 10 Most Common  Car brands

```python
brand_car = df['car'].value_counts().head(10)
plt.figure(figsize=(8, 6))
sns.barplot(x=brand_car.values, y=brand_car.index, palette="viridis")
plt.title("Top 10 Most Common Car Brands")
plt.xlabel("Number of Cars")
plt.ylabel("Car Brand")
plt.tight_layout()
plt.show()
```

Vechicle register or not

```python
# Count of registered vs non-registered cars
registration_counts = df['registration'].value_counts()
# Display counts
registration_counts
```

```python
# Plot
plt.figure(figsize=(6, 4))
sns.barplot(x=registration_counts.index, y=registration_counts.values)
plt.title("Registered vs Non-Registered Cars")
plt.xlabel("Registration Status")
plt.ylabel("Number of Cars")
plt.tight_layout()
plt.show()
```

```python
avg_price_by_brand = df.groupby('car')['price'].mean().sort_values(ascending=False).head(10)
print(avg_price_by_brand)
```

Average price of top 10 car

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_price_by_brand.values, y=avg_price_by_brand.index, palette="magma")
plt.title("Top 10 Car Brands by Average Price")
plt.xlabel("Average Price")
plt.ylabel("Car Brand")
plt.tight_layout()
plt.show()
```

Relation bw engV and price

```python
df_filter = df.dropna(subset=['engV', 'price'])

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_filter['engV'], y=df_filter['price'], color='red')
plt.title("Relation between engV and price")
plt.xlabel("engV")
plt.ylabel("price")
plt.tight_layout()
plt.show()
```

```python
df_fil = df.dropna(subset=['price', 'mileage'])
correlation = df_fil['mileage'].corr(df_fil['price'])
print(f"Correlation between mileage and price: {correlation:.2f}")
```

--> corr() gives the Pearson correlation coefficient, ranging from -1 to +1.

--> If close to -1, strong negative correlation (more mileage → less price).

--> regplot() fits and shows a regression line to highlight the trend.

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=df_fil, x='mileage', y='price', scatter_kws={'alpha':0.3}, line_kws={"color": "red"})
plt.title("Mileage vs Price with Regression Line")
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.tight_layout()
plt.show()
```

How many missing value in engV and drive

```python
missing_engV = df['engV'].isna().sum()
missing_drive = df['drive'].isna().sum()
print(missing_engV)
print(missing_drive)
```

```python
total_rows = len(df)
print(f"Percentage missing in 'engV': {missing_engV / total_rows * 100:.2f}%")
print(f"Percentage missing in 'drive': {missing_drive / total_rows * 100:.2f}%")
```

Check and replace missing value with mode and median

```python
df['engV'] = df['engV'].fillna(df['engV'].median())
df['drive'] = df['drive'].fillna(df['drive'].mode())
```
